```python
def barPosition(var, n_vars):
    separation =1
    width=0.8
    if n_vars%2==0:
        return ((var-n_vars/2)/n_vars)*width*separation
    else:
        return ((var-n_vars/2)/n_vars)*width*separation-width/(2*n_vars)
```

```python
barwidth =0.8/3
for i_gaussian in range(1, 10):
    index_data = index[f"gaussian_{i_gaussian}"]
    kraskov_data = kraskov[f"gaussian_{i_gaussian}"]
    copulas_data = copulas[f"gaussian_{i_gaussian}"]
    
    index_TC, index_DTC = np.array(index_data["TC_list"]), np.array(index_data["DTC_list"])
    kraskov_TC, kraskov_DTC = np.array(kraskov_data["TC_list"]), np.array(kraskov_data["DTC_list"])
    copulas_TC, copulas_DTC = np.array(copulas_data["TC_list"]), np.array(copulas_data["DTC_list"])
    
    TC_comp = abs(index_TC - kraskov_TC)
    DTC_comp = abs(index_DTC - kraskov_DTC)
    TC_comp_copulas = abs(index_TC - copulas_TC)
    DTC_comp_copulas = abs(index_DTC - copulas_DTC)
    
    fig, ax = plt.subplots(2, 2, figsize=(10,5))    
    
    # real vs methods
    ax[0, 0].bar(np.arange(len(index_TC))+barPosition(1, 3), index_TC, width=barwidth, label="real")
    ax[0, 0].bar(np.arange(len(kraskov_TC))+barPosition(2, 3), kraskov_TC, width=barwidth, label="kraskov")
    ax[0, 0].bar(np.arange(len(copulas_TC))+barPosition(3, 3), copulas_TC, width=barwidth, label="copulas")
    ax[0, 0].grid(axis='y')
    ax[0, 0].set_ylim([0, 13])
    ax[0, 0].legend()
    ax[0, 0].set_xticks([])
    ax[0, 0].set_ylabel("value")
    ax[0, 0].set_title("TC")
    ax[0, 1].bar(np.arange(len(index_DTC))+barPosition(1, 3), index_DTC, width=barwidth, label="real")
    ax[0, 1].bar(np.arange(len(kraskov_DTC))+barPosition(2, 3), kraskov_DTC, width=barwidth, label="kraskov")
    ax[0, 1].bar(np.arange(len(copulas_DTC))+barPosition(3, 3), copulas_DTC, width=barwidth, label="copulas")
    ax[0, 1].grid(axis='y')
    ax[0, 1].legend()
    ax[0, 1].set_ylim([0, 13])
    ax[0, 1].set_xticks([])
    ax[0, 1].set_title("DTC")
    
    # Error real vs methods
    TC_labels = [r"$I(X_{}; X^{})$".format('{'+str(i+2)+'}','{'+str(i+1)+'}') for i in range(len(TC_comp))]
    DTC_labels =[r"$I(X_{}; X^{})$".format('{'+str(len(TC_comp)+1)+'}','{'+str(len(TC_comp))+'}')] + [r'$I(X_{}; X^{}|X^{}_{})$'.format('{'+str(i+2)+'}', i+1 ,'{'+str(len(DTC_comp)+1)+'}', '{'+str(i+3)+'}') for i in range(len(DTC_comp)-1)] 

    ax[1, 0].bar(np.arange(len(TC_labels))+barPosition(1, 2), TC_comp/index_TC, width=0.4, label='kraskov', color='C1')
    ax[1, 0].bar(np.arange(len(TC_labels))+barPosition(2, 2), TC_comp_copulas/index_TC, width=0.4, label='copulas', color='C2')
    ax[1, 0].grid(axis='y')
    ax[1, 0].legend()
    ax[1, 0].set_xticks(np.arange(len(TC_labels))+0.2)
    ax[1, 0].set_xticklabels(TC_labels, rotation=45)
    ax[1, 0].set_ylabel("absolute error")
    ax[1, 0].set_ylim([0, 1.1])
    
    ax[1, 1].bar(np.arange(len(DTC_labels))+barPosition(1, 2), DTC_comp/index_DTC, width=0.4, label='kraskov', color='C1')
    ax[1, 1].bar(np.arange(len(DTC_labels))+barPosition(2, 2), DTC_comp_copulas/index_DTC, width=0.4, label='copulas', color='C2')
    ax[1, 1].grid(axis='y')
    ax[1, 1].legend()
    ax[1, 1].set_xticks(np.arange(len(DTC_labels))+0.2)
    ax[1, 1].set_xticklabels(DTC_labels, rotation=45)
    ax[1, 1].set_ylim([0, 1.1])
    
    plt.suptitle(f"gaussian_{i_gaussian}")
    
    plt.tight_layout()
    plt.show()
    fig.savefig(f'./plots/barras/bars{i_gaussian}.pdf', dpi=500)
```
